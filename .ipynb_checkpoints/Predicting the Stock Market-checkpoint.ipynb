## Predicting the Stock Market

In this project, we will work with data from the __[S&P500 Index](https://en.wikipedia.org/wiki/S%26P_500_Index)__. The S&P500 is a stock market index. Indexes aggregate the prices of multiple stocks together, and allow you to see how the market as a whole is performing. We will be using historical data on the price of the S&P500 Index to make predictions about future prices. Predicting whether an index will go up or down will help us forecast how the stock market as a whole will perform. Since stocks tend to correlate with how well the economy as a whole is performing, it can also help us make economic forecasts.

Note: You shouldn't make trades with any models developed in this project. Trading stocks has risks, and nothing in this project constitutes stock trading advice.

Each row in the file contains a daily record of the price of the S&P500 Index from 1950 to 2015. We will use this dataset to develop a predictive model. We will train the model with data from 1950-2012, and try to make predictions from 2013-2015.

### Summary of Results
Our Mean Absolute Error is off by just over 16 points and our average closing price is just under 1875. This isn't a model we would want to use to actually make stock trades, but it isn't bad considering we only used features derived from the closing price and volume history.

### Reading In the Data

```python
#Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
```

```python
#Read data from text file
data = pd.read_csv("stocks.txt")
```

```python
#Check data
data.head(2)
```

```python
#Convert the Date column to a Pandas date type to do date comparisons with the column
data["Date"] = pd.to_datetime(data["Date"])
```

```python
#Sort the dataframe on the Date column
data = data.sort_values(by = ["Date"])
data = data.reset_index(drop=True)
```

```python
#Check sorted data
data.head()
```

### Generating Indicators
Stock market data is sequential, and each observation comes a day after the previous observation. Thus, the observations are not all independent, and you can't treat them as such. Y

The time series nature of the data means that can generate indicators to make our model more accurate. ou have to be extra careful to not inject "future" knowledge into past rows when you do training and prediction. Injecting future knowledge will make our model look good when you're training and testing it, but will make it fail in the real world. When you do this, you have to be careful not to use the current row in the values you average. You want to teach the model how to predict the current price from historical prices.

We are going to generate below indicators for our prediction:
1. The average price from the past 5 days, 30 days, 360 days.
2. The standard deviation from the past 5 days, 30 days, 360 days.
3. The volume of the price over the past 5 days, 30 days, 360 days.
4. The average standard deviation from the past 5 days.
5. Minimum and maximum price in the past 365 days.
6. The ratio between the lowest price in the past year and the current price.
7. The ratio between the highest price in the past year and the current price.

#### Average price from the past 5 days, 30 days, 360 days
To compute indicators, we will loop through each day from 1951-01-03 to 2015-12-07. For instance, if we were computing the average price from the past 5 days, we will start at 1951-01-03, get the prices for each day from 1950-12-26 to 1951-01-02, and find the average. When we compute the average of the past 5 days for 1951-01-04, we don't include 1951-01-04 in that average.

```python
#Average price from past 5 days using the rolling function to compute the mean for the given window
data["past_5_avg"] = data["Close"].rolling(5).mean()

#Reindex the resulting series to shift all the values "forward" one day. 
#For example, the rolling mean calculated for 1950-01-03 will need to be assigned to 1950-01-04, and so on. 
data["past_5_avg"] = data["past_5_avg"].shift()

#Calculate average price from past 30 days
data["past_30_avg"] = data["Close"].rolling(30).mean()
data["past_30_avg"] = data["past_30_avg"].shift()

#Calculate average price from past 365 days
data["past_365_avg"] = data["Close"].rolling(365).mean()
data["past_365_avg"] = data["past_365_avg"].shift()
```

```python
#Check data
data.head()
```

```python
#Replace null values in the indicators we created with 0 because null values will be created during the computation
data.update(data[["past_5_avg", "past_30_avg", "past_365_avg"]].fillna(0))
```

```python
#Check data
data.head()
```

#### Standard Deviation from the past 5 days, 30 days, 360 days

```python
#Calculate average standard deviation for past 5 days
data["past_5_sd"] = data["Close"].rolling(5).std()
data["past_5_sd"] = data["past_5_sd"].shift()

#Calculate average standard deviation for past 30 days
data["past_30_sd"] = data["Close"].rolling(30).std()
data["past_30_sd"] = data["past_30_sd"].shift()

#Calculate average standard deviation for past 365 days
data["past_365_sd"] = data["Close"].rolling(365).std()
data["past_365_sd"] = data["past_365_sd"].shift()
```

```python
#Replace null values in the above columns with 0
data.update(data[["past_5_sd", "past_30_sd", "past_365_sd"]].fillna(0))
```

```python
#Check data
data.head(10)
```

#### Volume from the past 5 days, 30 days, 360 days

```python
#Calculate average volume for past 5 days
data["past_5_vol"] = data["Volume"].rolling(5).mean()
data["past_5_vol"] = data["past_5_vol"].shift()

#Calculate average volume for past 30 days
data["past_30_vol"] = data["Volume"].rolling(30).mean()
data["past_30_vol"] = data["past_30_vol"].shift()

#Calculate average volume for past 365 days
data["past_365_vol"] = data["Volume"].rolling(365).mean()
data["past_365_vol"] = data["past_365_vol"].shift()
```

```python
#Replace null values in the above columns with 0
data.update(data[["past_5_vol", "past_30_vol", "past_365_vol"]].fillna(0))
```

```python
#Check data
data.head(20)
```

#### The average standard deviation from the past 5 days

```python
data["past_5_avg_sd"] = data["past_5_avg"].rolling(5).std()
data["past_5_avg_sd"] = data["past_5_avg_sd"].shift()
```

```python
#Replace null values in the above columns with 0
data.update(data[["past_5_avg_sd"]].fillna(0))
```

```python
#Check data
data.head(20)
```

#### Minimum and maximum price in the past 365 days

```python
#Calculate the mimimum price in the past 365 days
data["past_365_min"] = data["Close"].rolling(365).min()
data["past_365_min"] = data["past_365_min"].shift()

#Calculate the maximum price in the past 365 days
data["past_365_max"] = data["Close"].rolling(365).max()
data["past_365_max"] = data["past_365_max"].shift()
```

```python
#Replace null values in the above columns with 0
data.update(data[["past_365_min", "past_365_max"]].fillna(0))
```

```python
#Check data
data.tail(10)
```

#### Ratio Between the Lowest and the Highest Price in the Past Year and the Current Price

```python
#Calculate ratio for minimum and maximum price in the past year
for index, row in data.iterrows():
    if index >= 365:
        i = index - 1
        data["ratio_price_365_min"] = data["Close"].iloc[i] / data["past_365_min"].iloc[index]
        data["ratio_price_365_max"] = data["Close"].iloc[i] / data["past_365_max"].iloc[index]
```

```python
#Eliminate rows with 0 in new columns and NA values
data = data.dropna(axis = 0)
```

#### Splitting up the data into train and test dataframes

```python
#Train data will contain any rows in the data with a date less than 2013-01-01
train = data[data["Date"] < pd.datetime(year=2013, month=1, day=1)]

#Test should contain any rows with a date greater than or equal to 2013-01-01
test = data[data["Date"] >= pd.datetime(year=2013, month=1, day=1)]
```

### Making Predictions
We will use Logistic Regression to predict the cost using different features/indicators. We will use the Mean Absolute Error (MAE) to check the error in prediction.

```python
#Using only past 5 days average price to predict closing price
lr = LinearRegression()
lr.fit(train[["past_5_avg"]], train["Close"])
predictions = lr.predict(test[["past_5_avg"]])
mae = mean_absolute_error(test["Close"], predictions)
print(mae)
```

### Improving Error

```python
#Using past 5, 30 and 365 days average price to predict closing price
lr = LinearRegression()
lr.fit(train[["past_5_avg", "past_30_avg", "past_365_avg"]], train["Close"])
predictions = lr.predict(test[["past_5_avg", "past_30_avg", "past_365_avg"]])
mae = mean_absolute_error(test["Close"], predictions)
print(mae)
```

```python
#Using past 5, 30 and 365 days average price and standard deviation to predict closing price 
lr = LinearRegression()
lr.fit(train[["past_5_avg", "past_30_avg", "past_365_avg", "past_5_sd", "past_30_sd", "past_365_sd"]], train["Close"])
predictions = lr.predict(test[["past_5_avg", "past_30_avg", "past_365_avg", "past_5_sd", "past_30_sd", "past_365_sd"]])
mae = mean_absolute_error(test["Close"], predictions)
print(mae)
```

```python
#Using all the indicators to predict closing price
features = ["past_5_avg", "past_30_avg", "past_365_avg", "past_5_sd", "past_30_sd", "past_365_sd", "past_5_vol", "past_30_vol", "past_365_vol", "past_5_avg_sd", "past_365_min", "past_365_max", "ratio_price_365_min", "ratio_price_365_max"]
lr = LinearRegression()
lr.fit(train[features], train["Close"])
predictions = lr.predict(test[features])
mae = mean_absolute_error(test["Close"], predictions)
print(mae)
```

Our Mean Absolute Error is off by just over 16 points and our average closing price is just under 1875. This isn't a model we would want to use to actually make stock trades, but it isn't bad considering we only used features derived from the closing price and volume history.

### Plot the Actual and Predicted Closing Prices

```python
close = test["Close"]
date = test["Date"]
fig = plt.figure(figsize=(10,5))
plt.plot(date, close, c='red')
plt.plot(date, predictions, c='blue')
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("ACTUAL vs. PREDICTED CLOSING PRICES")
plt.show()
```
